```python
# 1. Class & Object
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def show_details(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

s1 = Student("Prarav", 90)
s1.show_details()
```

```python
# 2. Encapsulation
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks

    def get_marks(self):
        return self.__marks

s = Student("Vivaan", 85)
print(s.get_marks())
```

```python
# 3. Inheritance
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)
        self.marks = marks

s = Student("Ananya", 78)
print(s.name, s.marks)
```

```python
# 4. Polymorphism
class CSVFile:
    def read(self):
        print("Reading CSV file")

class ExcelFile:
    def read(self):
        print("Reading Excel file")

def load_data(file):
    file.read()

load_data(CSVFile())
load_data(ExcelFile())
```

```python
# 5. Abstraction

from abc import ABC, abstractmethod

class DataSource(ABC):
    @abstractmethod
    def load(self):
        pass

class API(DataSource):
    def load(self):
        print("Loading data from API")

obj = API()
obj.load()
```

```python
# 6. Class and static methods
class DataHelper:
    source = "Kaggle"

    @classmethod
    def get_source(cls):
        return cls.source

    @staticmethod
    def clean_data():
        print("Cleaning data")

print(DataHelper.get_source())
DataHelper.clean_data()
```

```python
# 7. Mini Data Analysis Example
class StudentAnalysis:
    def __init__(self, data):
        self.data = data

    def get_average(self):
        total = 0

        for d in self.data:
            print(f"Adding marks of {d['name']}: {d['marks']}")
            total += d["marks"]

        print(f"Total Marks = {total}")
        avg = total / len(self.data)
        print(f"Number of Students = {len(self.data)}")

        return avg

data = [
    {"name": "Aarav", "marks": 90},
    {"name": "Riya", "marks": 80}
]

obj = StudentAnalysis(data)

result = obj.get_average()
print(f"Average Marks = {result}")
```
